```python
import string
import random
def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
  return ''.join(random.choice(chars) for _ in range(size))
```

```python
import torch
from torchvision import models
```

```python
dir(models)
```

```python
alexnet = models.alexnet(pretrained=True)
alexnet.eval()
```

```python
from torchvision import transforms

transform = transforms.Compose([            
transforms.Resize(256),                    
transforms.CenterCrop(224),                
transforms.ToTensor(),                     
transforms.Normalize(                      
mean=[0.485, 0.456, 0.406],                
std=[0.229, 0.224, 0.225]                  
 )])
```

```python
with open('/content/imagenet_classes.txt') as f:
  classes = [line.strip() for line in f.readlines()]
labels = classes
```

```python
import os
```

```python
from PIL import Image
```

```python
image = []
label = []
```

```python
for i in os.listdir('/content/content/files/val2017'):
  img = Image.open('/content/content/files/val2017/'+i)
  try:
    img_t = transform(img)
    batch_t = torch.unsqueeze(img_t, 0)
    out = alexnet(batch_t)
    _, index = torch.max(out, 1)
    percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100
    print(labels[index[0]], percentage[index[0]].item())
    image.append(i)
    label.append(labels[index[0]], percentage[index[0]].item())
  except:
    pass
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'images':image,'label':label})
```
